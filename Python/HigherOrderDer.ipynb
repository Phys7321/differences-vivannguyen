```python
import math
import numpy as np
from numpy import array
from myder import *
from matplotlib import pyplot as plt

dx = np.linspace(0,2*np.pi,99);
y = lambda x: np.power(np.sin(1/(x*(2-x))), 2)

y0 = np.power(np.sin(1/(dx*(2-dx))), 2)

p0 = plt.plot(dx,y0)
plt.show()
```

```python
dyf = forwardiff(y,2*np.pi,0,100)
dyb = backwardiff(y,2*np.pi,0,100)
dyc = centraldiff(y,2*np.pi,0,99)
dycubic = centraldiffcubic(y,2*np.pi,0,99)

print (2*np.pi)

print (dx)
print (dyf)

p1, = plt.plot(dx, dyf, label='forward')
p2, = plt.plot(dx, dyb, label='backward')
p3, = plt.plot(dx, dyc, label='central')
p4, = plt.plot(dx, dycubic, label='central cubic')
plt.legend([p1, p2, p3, p4])
plt.show()
```
